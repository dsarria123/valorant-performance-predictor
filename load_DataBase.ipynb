### Load database(insert ids of players)

```python
import pandas as pd
from Scraping.MatchScraperCopy import scrape_player_dataCopy
from Preprocessing.preprocessData import preprocess_data
from databaseFunctions import load_database

# Put player id here
daplayeridiots = []   

OPPOSING_TEAM = "Sentinels"
TABLE_NAME = 'player_matches'

for player_id in daplayeridiots:
    print(f"Scraping data for Player ID: {player_id}...")

    raw_df = scrape_player_dataCopy(player_id, OPPOSING_TEAM)
    if raw_df is None or raw_df.empty:
        print(f"No data found for Player ID {player_id}. Skipping.")
        continue

    officialdf = preprocess_data(raw_df)
    load_database(officialdf, TABLE_NAME)

    officialdf.to_csv(f"{player_id}.csv", index=False)
    print(f"Saved: {player_id}.csv\n")
```

### Transfer data between databases(Done but will leave here in case need again)

```python
import pandas as pd
from sqlalchemy import create_engine

# Engine for local DB
local_engine = create_engine("")

# Engine for Supabase (transaction pooler)
supabase_engine = create_engine(
    "",
    pool_pre_ping=True
)

# Step 1: Load data from local DB
df = pd.read_sql("SELECT * FROM player_matches", local_engine)

# Step 2: Upload to Supabase
df.to_sql("player_matches", supabase_engine, if_exists="append", index=False)

print(f"Transferred {len(df)} rows from local DB to Supabase.")
```
